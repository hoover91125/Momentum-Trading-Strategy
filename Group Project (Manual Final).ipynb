```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
ticker=pd.read_csv('dowjones.csv')
```

```python
#Ticker of Dow Jones component from 2010 to 2020
ticker
```

```python
#2010 daily adj close price for each components
df10 = pdr.get_data_yahoo(ticker.iloc[0:29,0], start='2010-01-01',end='2010-12-31')['Adj Close']
df10
```

```python
#drop stocks w/ data
df10.drop(labels=['HPE','UTX'],axis='columns',inplace=True)
df10
```

```python
#simple return in 2010
df10_return=df10.pct_change()
df10_return
```

```python
#2010 average daily return
df10_avg_dailyreturn=df10_return.mean()
df10_avg_dailyreturn
```

```python
#2010 average annual return
df10_avg_annualreturn=df10_return.mean()*250
df10_avg_annualreturn
```

```python
#sort annual return in decending order
df10_avg_annualreturn.sort_values(inplace=True, ascending=False)
df10_avg_annualreturn
```

```python
#select the top 10 stocks
r10_selected=df10_avg_annualreturn[0:10]
r10_selected
```

```python
Dowjones = pdr.get_data_yahoo('^DJI', start='2011-01-01',end='2020-12-31')
Dowjones
```

```python
Dowjones['Dow Jones Simple Return']=Dowjones['Adj Close'].pct_change()
Dowjones
```

```python
Dowjones.drop(labels=['High', 'Low','Open','Close','Volume'],axis='columns',inplace=True)
Dowjones
```

```python
Dowjones['Year']=Dowjones.index.year
Dowjones
```

```python
Dowjones.drop(labels=['Adj Close'],axis='columns',inplace=True)
Dowjones
```

```python
R_Dow=Dowjones.groupby('Year').mean()*250
R_Dow
```

```python
port=[port_11,port_12,port_13,port_14,port_15,port_16,port_17,port_18,port_19,port_20 ]
R_Dow['Portfolio Return']=port
R_Dow
```

```python
R_Dow[['Dow Jones Simple Return','Portfolio Return']].plot(figsize=(20, 10),  marker='o',markersize=10.0,linewidth=4.0)
plt.xlabel('Year', fontsize=18)
plt.title('Dow Jones Simeple Return vs. Portfolio Simple Return', fontsize=18)
plt.legend(prop={'size': 20})
plt.grid(True)
plt.xticks(R_Dow.index) 
plt.show()
```
